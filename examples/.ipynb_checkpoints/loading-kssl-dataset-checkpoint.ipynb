* **Count number of measurements by analyte**

**Note:** 
* Only measurements with associated MIRS scan
* Here below in 4 steps but can be done in 1!

```python
# Step 1: Laoding fact table (selecting only relevant columns)
df_fact = load_fact_tbl().loc[:, ['lay_id', 'lims_pedon_id', 
                                  'lay_depth_to_top', 'smp_id', 'analyte_id']]

df_fact.head()
```

```python
# Step 2: Load analytes (and relevant columns)
df_analytes = load_analytes()[['analyte_id', 'analyte_name']]
```

```python
# Step 2: Load spectrum (keeping only their smp_id)
df_spectra = load_spectra().iloc[:, [0]]
```

```python
# Step 3: Join/merge tables
df = pd.merge(df_fact, df_spectra, on='smp_id')
```

```python
# Step 4: Count by group
df_count = df.groupby('analyte_id') \
    .count()[['lay_id']] \
    .rename(columns={'lay_id': 'dataset_size'}) \
    .sort_values(by='dataset_size', ascending=False) \
    .merge(df_analytes, on='analyte_id')

df_count.head(15)
```

```python
# Getting analyte_id of analyte measured on more than X samples
mask_not_mirs_predicted = ~df_count['analyte_name'].str.contains('MIR predicted')
mask_over_X = df_count['dataset_size'] > 30000

top_measured_analytes = df_count.loc[mask_over_X & mask_not_mirs_predicted, 'analyte_id'].to_list()
top_measured_analytes[:10] # First ten
```

```python
# In case export is required (here as an Excel file)
df_count.to_excel(__YOUR_PATH__)
```
