```python
DATA_PATH = Path.home() / 'data_spectrai/kssl'
for f in DATA_PATH.glob('*.csv'):
    print(f)
```

```python
props_in_tables = {}
#for f in glob.glob(FILES_PATH):
for f in DATA_PATH.glob('*.csv'):
    f_base = os.path.basename(f)
    df = pd.read_csv(f)
    for col in list(df.columns):
        if col not in props_in_tables:
            props_in_tables[col] = [f_base]
        else: 
            props_in_tables[col].append(f_base)
```

```python
props_in_tables
```

```python
# Creating graph of table's relation
nodes = list(map(lambda f: os.path.basename(f), glob.glob(FILES_PATH)))

MG = nx.MultiGraph()
MG.add_nodes_from(nodes, shape='circle', style='filled', fillcolor='#333333', fontcolor='white', fontsize=50)

for attr, files in props_in_tables.items():
    file_pairs = list(set(itertools.combinations(files, 2)))
    for p in file_pairs:
        MG.add_edge(*p, label=attr, color='#333333', fontsize=50)  

A = to_agraph(MG)
A.layout('dot')
A.draw('img/kssl-data-model.png')
```
