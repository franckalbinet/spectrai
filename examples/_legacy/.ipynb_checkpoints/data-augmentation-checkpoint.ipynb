## Utilities functions

```python
def get_range(X, axis=0, rounding=True, reversing=False):
    val_range = (np.amin(X, axis=axis), np.amax(X, axis=axis))
    if rounding==True:
        val_range = list(map(np.round, val_range))
    if reversing==True:
        return val_range[::-1]
    else:
        return val_range

def plot_spectra(X, labels, figsize=(15,8), reversing=True, xlabel='wavenumber'):
    low, high = get_range(labels, reversing=reversing)
    with plt.style.context(('ggplot')):
        fig, ax = plt.subplots(figsize=figsize)
        ax.set_xlim(low, high)
        plt.xlabel(xlabel)
        _ = ax.plot(labels, X.T)
        
def standardize(X):
    return (X - X.mean()) - X.std()

def normalize(X):
    return (X - X.min()) / (X.max() - X.min())
```

## Basic Variational Auto-encoder (VAE)

```python
codings_size = 30

input_dim = X.shape[1]
print("input_dim: ", input_dim)

class Sampling(layers.Layer):
    def call(self, inputs):
        mean, log_var = inputs
        return K.random_normal(tf.shape(log_var)) * K.exp(log_var / 2) + mean 
```

```python
# Encoder
inputs = layers.Input(shape=(input_dim,))
z = layers.Dense(150, activation='relu')(inputs)
z = layers.Dense(10, activation='relu')(z)
codings_mean = layers.Dense(codings_size)(z)
codings_log_var = layers.Dense(codings_size)(z)
codings = Sampling()([codings_mean, codings_log_var])
variational_encoder = Model(inputs = [inputs], outputs=[codings_mean, codings_log_var, codings])

# Decoder
decoder_inputs = layers.Input(shape=[codings_size])
x = layers.Dense(100, activation="selu")(decoder_inputs)
x = layers.Dense(150, activation="selu")(x)
outputs = keras.layers.Dense(input_dim, activation="linear")(x)
#outputs = keras.layers.Reshape([28, 28])(x)
#x = keras.layers.Dense(28 * 28, activation="sigmoid")(x)
#outputs = keras.layers.Reshape([28, 28])(x)
variational_decoder = keras.models.Model(inputs=[decoder_inputs], outputs=[outputs])

# Build VAE model
_, _, codings = variational_encoder(inputs)
reconstructions = variational_decoder(codings)
variational_ae = keras.models.Model(inputs=[inputs], outputs=[reconstructions])
```

```python
# When not trained: looking at the vae output
fig, ax = plt.subplots(figsize=(15,8))
_ = ax.plot(X_names, variational_ae(standardize(X)).numpy().T)
```

```python
# Training
latent_loss = -0.5 * K.sum(
    1 + codings_log_var - K.exp(codings_log_var) - K.square(codings_mean),
    axis=-1)

#variational_ae.add_loss(K.mean(latent_loss) / 784.)
variational_ae.add_loss(K.mean(latent_loss) / float(input_dim))
                        
#variational_ae.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=[rounded_accuracy])
variational_ae.compile(loss="mse", optimizer="rmsprop")

#history = variational_ae.fit(X_train, X_train, epochs=25, batch_size=128,
#                             validation_data=[X_valid, X_valid])

history = variational_ae.fit(X, X, epochs=10000)
```

## CNN Variational Auto-encoder (VAE)

```python
codings_size = 10
activation = 'relu'
padding = 'same'

#input_dim = X.shape[1]
input_dim = 1800

print("input_dim: ", input_dim)

class Sampling(layers.Layer):
    def call(self, inputs):
        mean, log_var = inputs
        return K.random_normal(tf.shape(log_var)) * K.exp(log_var / 2) + mean 
```

```python
# Encoder
inputs = layers.Input(shape=(input_dim,))
z = layers.Reshape((input_dim, 1), input_shape=(input_dim,))(inputs)

z = layers.Conv1D(32, 30, activation=activation, padding=padding)(z)
z = layers.MaxPool1D(10)(z)

z = layers.Conv1D(64, 30, activation=activation, padding=padding)(z)
z = layers.MaxPool1D(10)(z)

z = layers.Flatten()(z)
z = layers.Dense(100, activation=activation)(z)
z = layers.Dense(50, activation=activation)(z)

codings_mean = layers.Dense(codings_size)(z)
codings_log_var = layers.Dense(codings_size)(z)
codings = Sampling()([codings_mean, codings_log_var])
variational_encoder = Model(inputs = [inputs], outputs=[codings_mean, codings_log_var, codings])
```

```python
# Decoder
decoder_inputs = layers.Input(shape=[codings_size])

x = layers.Dense(50, activation=activation)(decoder_inputs)
x = layers.Dense(100, activation=activation)(x)
x = layers.Dense(1152, activation=activation)(x)
x = layers.Reshape(target_shape=(18, 64))(x) # To get last encoder's pooling layer output size automatically instead

x = layers.Conv1D(64, 30, activation=activation, padding=padding)(x)
x = layers.UpSampling1D(10)(x)

x = layers.Conv1D(32, 30, activation=activation, padding=padding)(x)
x = layers.UpSampling1D(10)(x)

x = layers.Conv1D(1, 30, activation='sigmoid', padding=padding)(x)
variational_decoder = keras.models.Model(inputs=[decoder_inputs], outputs=[x])
```

## Altering baseline offset and slope

Instead of correction of the baseline variations we use a special implementation of data augmentation. The idea is to simulate the expected form of irrelevant noise (here baseline offset and slope and overall spectrum intensity), and expect the neural network to either extract features robust to the variations, or figure out the best corrections during training. This way we can make more efficient use of the labelled dataset. The idea is similar to the image rotations and croppings done in image recognition training.

Source: https://github.com/EBjerrum/Deep-Chemometrics/blob/master/Deep_Chemometrics_with_data_augmentation.py.ipynb

```python
def augment_data(X, y, by=10, betashift = 0.05, slopeshift = 0.05, multishift = 0.05):
    X_aug = np.repeat(X, repeats=by, axis=0)
    X_aug = _alter(X_aug, betashift = betashift, slopeshift = slopeshift, multishift = multishift)
    y_aug = np.repeat(y, repeats=by, axis=0) #y_train is simply repeated
    return (X_aug, y_aug)
```

```python
def _alter(x, betashift = 0.05, slopeshift = 0.05, multishift = 0.05):
    # Shift of baseline
    samples_unif = np.random.random(size=(x.shape[0],1))  # From continuous uniform [O,1) interval dist.
    beta = betashift * (2 * samples_unif - 1) # -betashift <= beta < betashift
    slope = slopeshift * (2 * slopeshift - 1) + 1 # 1 - slopeshift <= slope < 1 + slopeshift
    
    # Calculate relative position
    axis = np.array(range(x.shape[1]))/float(x.shape[1])
    
    # Calculate offset to be added
    offset = slope*(axis) + beta - axis - slope/2. + 0.5
    
    # Multiplicative
    multi = np.random.random(size=(x.shape[0],1))*2*multishift-multishift + 1
    
    return  multi*x + offset
```

```python
shift = 0.1
X_aug, y_aug = augment_data(X, y, by=10, betashift=shift, multishift=shift)
```
